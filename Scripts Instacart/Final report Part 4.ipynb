```python
#Panda libraries 

import pandas as pd
import numpy as np
import os

#Viz libraries 

import matplotlib.pyplot as plt
import seaborn as sns
import scipy

```

```python
#Importing the dataset 

path = r'/Users/alexbor/Desktop/Data Analytics/Immersion/Submissions/Instacart Basket Analysis'

ords_prods_cust_merge = pd.read_pickle(os.path.join(path, '02 Data ', 'Prepared data', 'ords_prods_cust_merge_4.10.pkl'))
```

# 6. Food preferences by unique user_id 

```python
unique_food_per_user_id = ords_prods_cust_merge.groupby('Food preferences')['user_id'].nunique()
```

```python
unique_food_per_user_id
```

```python
#Bar chart for the dataframe "unique_food_per_user_id"

unique_food_per_user_id_bar = unique_food_per_user_id.plot.bar(color = ['red', 'yellow', 'orange','blue'])

plt.ylabel('Count of unique user_ids')
plt.title('Food preferences',fontsize=12)
plt.show()
```
