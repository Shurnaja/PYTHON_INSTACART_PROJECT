```python
#Import df 

#Create a path

path = r'/Users/alexbor/Desktop/Data Analytics/Immersion/Submissions/Instacart Basket Analysis'

#Import the df 

df_orders_products_merged = pd.read_pickle(os.path.join(path, '02 Data ', 'Prepared data', 'orders_products_merged_finall.pkl'))
```

```python
#Create a subset to reduce memory 

df = df_orders_products_merged[:1000000]
```

```python
#Defining the functions 

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df.shape
```

```python
df = df.drop(columns = ['Unnamed: 0'])
```

```python
df.shape
```

```python
#Apply the function 

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
#Flagging the days using for-loop 

result = []

for value in df_orders_products_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Printing the result
result
```

```python
#Creating a column with the results of for-loop

df_orders_products_merged["busiest day"]=result
```

```python
#Step 2. Create a new label for TWO busiest days "Busiest days". For TWO slowest days - new label "Least busy days"
#Write a for-loop code

result = []

for value in df_orders_products_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 1:
    result.append("Busiest day")
  elif value == 3:
    result.append("Least busy")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Create a new column in my df and put the "result" values in it
df_orders_products_merged["busiest day"]=result
```

```python
#Step 3. Check the value counts in the column "busiest day"
#Sat (0) and Sun (1) values are now labeled as "Busiest day"; Tue (3) and Wed (4) - as "Least busy"

df_orders_products_merged["busiest day"].value_counts(dropna=False)
```

```python
#Step 4. Identify busiest hours of day (orders_time)
#Check how busy each hours of the day is by counting their value frequencies 

df_orders_products_merged['orders_time'].value_counts(dropna=False)
```

```python
#Divide the hours of the day into labels “Most orders,” “Average orders,” and “Fewest orders.”
#I decided to divide the result column into 3 blocks with 8 items in each. 
#Write a for-loop code for it in separate lines

result = []

for value in df_orders_products_merged['orders_time']:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result.append('Most orders')
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result.append('Fewest orders')
  else:
    result.append('Average orders')
```

```python
#Create a column 'busiest_period_of_day'for result

df_orders_products_merged['busiest_period_of_day']=result
```

```python
#Check the value frequencies in the new column

df_orders_products_merged['busiest_period_of_day'].value_counts(dropna=False)
```
