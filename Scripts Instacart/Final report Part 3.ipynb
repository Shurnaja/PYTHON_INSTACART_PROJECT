```python
#Panda libraries 

import pandas as pd
import numpy as np
import os

#Viz libraries 

import matplotlib.pyplot as plt
import seaborn as sns
import scipy

```

```python
#Importing the dataset 

path = r'/Users/alexbor/Desktop/Data Analytics/Immersion/Submissions/Instacart Basket Analysis'

ords_prods_cust_merge = pd.read_pickle(os.path.join(path, '02 Data ', 'Prepared data', 'ords_prods_cust_merge_4.10.pkl'))
```

# 1. Customer age groups by unique user_ids

```python
#Grouping the Customer age groups by unique user_ids
unique_age_groups_per_user_id = ords_prods_cust_merge.groupby('Customer age group')['user_id'].nunique()
```

```python
#Result check 
unique_age_groups_per_user_id
```

```python
#Bar chart for the dataframe "Customer age groups by unique user_ids"

unique_age_groups_per_user_id_bar = unique_age_groups_per_user_id.plot.bar(color = ['green', 'yellow', 'orange','blue'])

plt.ylabel('Count of unique user_ids')
plt.title('Customer age groups',fontsize=12)
plt.show()
```

```python
# Exporting the viz Age 

unique_age_groups_per_user_id_bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'age_customers_unique_user_id.png'))
```
